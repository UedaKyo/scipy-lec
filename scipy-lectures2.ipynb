## 1.4.3 Figures,Subplots,Axis,Ticks
### 1.4.3.1 Figures
- Figuresとは "Figure #" をタイトルに持つウィンドウ  
1から番号が始まり複数のパラメータがある

- num;数字の数(デフォルト:1)
- figsize:インチ単位の幅、高さの図形サイズ(figure,figsize)
- dpi:1インチ当たりのドット数で表した解像度(figure.dpi)
- facecoler:描画背景の色(figure.facecoler)
- edgecoler:描画背景の周囲のエッジの色(figure.edgecolor)
- frameon:図枠を書くかどうか(True)

GUIを利用していれば右上のxで消せるが、closeを呼び出して閉じることができる  
1:現在のfigure 2:特定のfigure 3:全てのfigure

```python
plt.close(1)  #figure 1 を閉じる
```
